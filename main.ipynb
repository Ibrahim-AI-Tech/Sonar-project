```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Collect Data and Process It

```python
#loading Data
Sonar = pd.read_csv('sonar data.csv', header=None)
"""head tom print the first data and tail to print the last data"""
# Sonar.head(7)
# Sonar.tail(7)
"""to know number of rows and columns"""
#Sonar.shape
"""to know more info about the data"""
Sonar.describe()  #mean std variance min max  => Statistical Measurement
# Sonar.info()

print(Sonar.describe(),Sonar.info())
```

```python
#to know how many future in the CSV file
Sonar[60].value_counts()
```

M => Mine

R => Rock

```python
""" 60 is the numer of Columns that have the future ( R,M )"""
Sonar.groupby(60).mean()
```

```python
""" Separate Data """
X = Sonar.drop(columns=60, axis=1)
Y = Sonar[60]
print(X.shape,Y.shape) 
```

Train and Test Data

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, shuffle=True)

# x_train.shape
# x_test.shape
# y_train.shape
# y_test.shape
```

Training the model => Logistic Regression

```python
model = LogisticRegression()

"""Train the model"""

model.fit(x_train, y_train)
```

Model Evaluation

```python
""" accuracy in train data"""

x_train_prediction = model.predict(x_train)
# print(x_prediction)
accuracy_train = accuracy_score(y_train, x_train_prediction)
print(accuracy_train)
```

```python
""" accuracy in test data"""
x_test_prediction = model.predict(x_test)
accuracy_test = accuracy_score(y_test, x_test_prediction)
print(accuracy_test)
```
